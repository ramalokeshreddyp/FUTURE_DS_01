# Business Sales Performance Analytics
### Data Science & Analytics – Task 1 (2026)

This notebook analyzes business sales data to identify:
- Revenue trends over time
- Top-selling products
- High-value categories
- Regional performance

```python
from pathlib import Path
import sys
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')

PROJECT_ROOT = Path.cwd()
if (PROJECT_ROOT / 'src').exists() is False:
    PROJECT_ROOT = PROJECT_ROOT.parent

SRC_PATH = PROJECT_ROOT / 'src'
if str(SRC_PATH) not in sys.path:
    sys.path.append(str(SRC_PATH))

from analysis_utils import load_sales_data, build_kpis, monthly_revenue, top_products, category_performance, regional_performance
```

```python
data_path = PROJECT_ROOT / 'data' / 'processed' / 'business_sales_curated.csv'
df = load_sales_data(data_path)
df.head()
```

```python
kpi = build_kpis(df)
pd.DataFrame([kpi]).style.format({
    'total_revenue': '${:,.0f}',
    'total_profit': '${:,.0f}',
    'avg_order_value': '${:,.2f}',
    'profit_margin_pct': '{:.2f}%',
    'return_rate_pct': '{:.2f}%'
})
```

## Revenue Trends Over Time

```python
monthly = monthly_revenue(df)
plt.figure(figsize=(12, 5))
plt.plot(monthly['month_start'], monthly['revenue'], marker='o', label='Revenue')
plt.plot(monthly['month_start'], monthly['profit'], marker='o', label='Profit')
plt.title('Monthly Revenue and Profit Trend')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

## Top-Selling Products

```python
top_prod = top_products(df[df['revenue'] > 0], n=10)
top_prod
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=top_prod.sort_values('revenue', ascending=False), x='revenue', y='product_name')
plt.title('Top 10 Products by Revenue')
plt.xlabel('Revenue')
plt.ylabel('Product')
plt.tight_layout()
plt.show()
```

## Category and Regional Performance

```python
cat_perf = category_performance(df)
reg_perf = regional_performance(df)

fig, axes = plt.subplots(1, 2, figsize=(15, 5))
sns.barplot(data=cat_perf, x='category', y='revenue', ax=axes[0])
axes[0].set_title('Revenue by Category')
axes[0].tick_params(axis='x', rotation=35)

region_summary = reg_perf.groupby('region', as_index=False).agg(revenue=('revenue', 'sum'))
sns.barplot(data=region_summary, x='region', y='revenue', ax=axes[1])
axes[1].set_title('Revenue by Region')
axes[1].tick_params(axis='x', rotation=20)

plt.tight_layout()
plt.show()
```
